```python
import pandas as pd
import plotly.express as px

# 1) view_count numeric ga aylantirish va NaN -> 0 (agar kerak bo'lsa)
df['view_count'] = pd.to_numeric(df['view_count'], errors='coerce').fillna(0).astype(int)

# 2) eng mashhurlarni tanlash
top10 = df.sort_values('view_count', ascending=False).head(10).copy()

# 3) uzun sarlavhalarni qisqartirish (ko'rsatish uchun)
def short_title(t, n=40):
    t = str(t)
    return (t[:n-3] + '...') if len(t) > n else t

top10['title_short'] = top10['title'].apply(lambda x: short_title(x, 40))

# 4) plot (bar)
fig = px.bar(
    top10,
    x='title_short',
    y='view_count',
    hover_data=['title', 'download_count', 'vote_count', 'owner_name'],
    text='view_count',
    labels={'title_short': 'Title', 'view_count': 'View count'},
    title='Top 10 eng mashhur Kaggle datasetlari (view_count bo‘yicha)'
)

# 5) biroz formatlash
fig.update_traces(texttemplate='%{text:,}', textposition='outside')
fig.update_layout(
    xaxis_tickangle=-40,
    yaxis=dict(tickformat=','),
    margin=dict(t=80, b=150),
    uniformtext_minsize=8
)

fig.show()
```
